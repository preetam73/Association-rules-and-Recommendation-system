```python
user_book_df1 = book_df1.pivot_table(index='User_ID',
                                 columns='Book_Title',
                                 values='Book_Rating')
```

```python
user_book_df1
```

```python
user_book_df1.index = book_df1.User_ID.unique()
```

```python

user_book_df1.fillna(0, inplace=True)
```

#### 
calculating similarity between users

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_book_df1.values,metric = 'cosine')
```

```python
user_sim
```

```python
user_sim.shape
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
user_sim_df.index = book_df1.User_ID.unique()
user_sim_df.columns = book_df1.User_ID.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```
